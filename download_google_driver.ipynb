````python


import os
import shutil

# Đường dẫn nguồn (thư mục cần sao chép)
source_folder = "/content/drive/MyDrive/N4"

# Kiểm tra xem thư mục nguồn có tồn tại không
if not os.path.exists(source_folder):
    print(f"Lỗi: Không tìm thấy thư mục '{source_folder}'")
    print("Vui lòng kiểm tra lại đường dẫn và đảm bảo đã kết nối đúng Google Drive")
else:
    # Tạo thư mục đích để sao chép các file
    destination_folder = "/content/downloaded_folder_N4"
    if os.path.exists(destination_folder):
        # Xóa thư mục đích nếu đã tồn tại để tránh lỗi
        shutil.rmtree(destination_folder)

    os.makedirs(destination_folder)

    # Đếm số lượng file để theo dõi tiến trình
    total_files = sum([len(files) for _, _, files in os.walk(source_folder)])
    print(f"Tổng số file cần sao chép: {total_files}")

    # Sao chép tất cả files từ folder nguồn đến folder đích
    try:
        # Hiển thị nội dung thư mục nguồn

        print("\nNội dung thư mục nguồn:")
        for root, dirs, files in os.walk(source_folder):
            level = root.replace(source_folder, '').count(os.sep)
            indent = ' ' * 4 * level
            print(f"{indent}{os.path.basename(root)}/")
            sub_indent = ' ' * 4 * (level + 1)
            for f in files:
                print(f"{sub_indent}{f}")

        # Sao chép toàn bộ thư mục
        shutil.copytree(source_folder, destination_folder, dirs_exist_ok=True)

        print(f"\nĐã sao chép tất cả files từ '{source_folder}' sang '{destination_folder}'")

        # Nén thư mục để dễ dàng tải xuống
        zip_name = "downloaded_folder_N4.zip"
        !zip -r {zip_name} {destination_folder}
        print(f"Đã tạo file {zip_name} để tải xuống")

        # Hiển thị kích thước của file zip
        !du -h {zip_name}

        print("\nBạn có thể tải xuống file zip từ panel Files ở bên trái của Colab")

    except Exception as e:
        print(f"Có lỗi khi sao chép files: {str(e)}")
````
